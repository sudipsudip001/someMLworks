## Output parser

```python
import json
from langchain_core.output_parsers import BaseOutputParser

class LLMOutputToJsonParser(BaseOutputParser):
    """Parse the output of an LLM call to a JSON format."""

    def parse(self, text: str):
        """Parse the output on an LLM call into JSON."""
        lines = text.split("\n")
        countries = [line.split('. ')[1] for line in lines if line.strip()]

        output_dict = {f"country{i+1}": country for i, country in enumerate(countries)}

        json_output = json.dumps(output_dict, indent=2)
        return json_output

llm_output = "1. France\n2. Italy\n3. Spain\n4. Germany\n5. Portugal\n6. Argentina\n7. Chile\n8. United States\n9. Australia\n10. South Africa"

json_parser = LLMOutputToJsonParser()

output_json =  json_parser.parse(llm_output)
print(output_json)
```

## LCEL (LangChain Expression Language)

```python
import json
from langchain_core.output_parsers import BaseOutputParser
from langchain_core.prompts import PromptTemplate
import json

class LLMOutputToJsonParser(BaseOutputParser):
    """Parse the output of an LLM call to a JSON format"""

    def parse(self, text: str):
        """Parse the output on an LLM into JSON."""
        countries = []
        lines = text.split("\n")

        for line in lines:
            parts = line.split('. ')
            if len(parts) > 1:
                country_name = parts[1]
                countries.append(country_name)
            
        output_dict = {f"country{i+1}": country for i, country in enumerate(countries)}

        json_output = json.dumps(output_dict, indent=2)
        return json_output

model = GoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GOOGLE_API_KEY)
prompt = PromptTemplate.from_template(
    "List exactly 10 countries that import {product}. "
    "Respond ONLY with numbered list like:\n"
    "1. Country\n"
    "2. Country\n"
    "No extra text, no intro, no explanation."
)
chain = prompt | model | LLMOutputToJsonParser()

output = chain.invoke({"product": "rice"})
print(output)
```

## Sequential Chains

```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough

llm = GoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GOOGLE_API_KEY)

first_prompt = ChatPromptTemplate.from_template(
    "What is the best name to describe a company that makes {product}? Return only the name, no explanation."
)

second_prompt = ChatPromptTemplate.from_template(
    "Write a 20 words description for the following company: {company_name}"
)

chain_one = first_prompt | llm
chain_two = second_prompt | llm

overall_chain = (
    {"company_name": chain_one}
    | RunnablePassthrough.assign(description=chain_two)
)

result = overall_chain.invoke({"product": "gaming laptop"})
print("Company name: ", result["company_name"])
print("Description: ", result["description"])
```

## Router chains

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough, RunnableBranch
from langchain_google_genai import GoogleGenerativeAI

llm = GoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GOOGLE_API_KEY, temperature=0.7)

physics_prompt = PromptTemplate.from_template(
    "You are a physics expert. Answer this physics question: {input}"
)

math_prompt = PromptTemplate.from_template(
    "You are a math expert. Answer this math question: {input}"
)

default_prompt = PromptTemplate.from_template(
    "Answer this question: {input}"
)

router_prompt = PromptTemplate.from_template(
    "Classify the topic of this question as one word - either 'physics', 'math' or 'other'.\n"
)

router_chain = router_prompt | llm

branch = RunnableBranch(
    (lambda x: "physics" in x["topic"].lower(), physics_prompt | llm),
    (lambda x: "math" in x["topic"].lower(), math_prompt | llm),
    default_prompt | llm
)

full_chain = (
    RunnablePassthrough.assign(topic=router_chain)
    | branch
)

result = full_chain.invoke({"input": "What is the speed of light?"})
print(result)
```

## Few shot prompting in langchain

```python
from langchain_core.prompts import FewShotPromptTemplate, PromptTemplate

examples = [
    {
        "prompt": "This movie is amazing!",
        "label": "Positive"
    },
    {
        "prompt": "I didn't like the service provided.",
        "label": "Negative"
    },
    {
        "prompt": "The product exceeds my expectations!",
        "label": "Positive"
    },
    {
        "prompt": "The weather ruined our plans.",
        "label": "Negative"
    }
]
```

```python
example_prompt = PromptTemplate(input_variables=["prompt", "answer"], template="Prompt: {prompt}\n{label}")

print(example_prompt.format(**examples[0]))
```

```python
prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    suffix="prompt: {input}",
    input_variables=["input"]
)

print(prompt.format(input="I hate this food"))
```

```python
from langchain_google_genai import GoogleGenerativeAI

model = GoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GOOGLE_API_KEY)
chain = prompt | llm
chain.invoke("I hate this food!")
```

## RAG(Retrieval Augmented Generation)

```python
import json
from fpdf import FPDF
```

```python
json_data = {
    "name": "John Doe",
    "age": 30,
    "email": "johnny@email.com"
}

pdf_data = "This is a very small PDF file and you're welcome to opening it anytime you like."

html_data = "<html><body><h1>Hello, HTML!</h1><p>This is a small HTML file.</p></body></html>"

with open('data.json', 'w') as json_file:
    json.dump(json_data, json_file)

pdf = FPDF()
pdf.add_page()
pdf.set_font('Arial', size=12)
pdf.cell(200, 10, txt=pdf_data, ln=True)
pdf.output('data.pdf')

with open('data.html', 'w') as f:
    f.write(html_data)
```
