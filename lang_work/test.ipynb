## Output parser

```python
import json
from langchain_core.output_parsers import BaseOutputParser

class LLMOutputToJsonParser(BaseOutputParser):
    """Parse the output of an LLM call to a JSON format."""

    def parse(self, text: str):
        """Parse the output on an LLM call into JSON."""
        lines = text.split("\n")
        countries = [line.split('. ')[1] for line in lines if line.strip()]

        output_dict = {f"country{i+1}": country for i, country in enumerate(countries)}

        json_output = json.dumps(output_dict, indent=2)
        return json_output

llm_output = "1. France\n2. Italy\n3. Spain\n4. Germany\n5. Portugal\n6. Argentina\n7. Chile\n8. United States\n9. Australia\n10. South Africa"

json_parser = LLMOutputToJsonParser()

output_json =  json_parser.parse(llm_output)
print(output_json)
```

## LCEL (LangChain Expression Language)

```python
import json
from langchain_core.output_parsers import BaseOutputParser
from langchain_core.prompts import PromptTemplate
import json

class LLMOutputToJsonParser(BaseOutputParser):
    """Parse the output of an LLM call to a JSON format"""

    def parse(self, text: str):
        """Parse the output on an LLM into JSON."""
        countries = []
        lines = text.split("\n")

        for line in lines:
            parts = line.split('. ')
            if len(parts) > 1:
                country_name = parts[1]
                countries.append(country_name)
            
        output_dict = {f"country{i+1}": country for i, country in enumerate(countries)}

        json_output = json.dumps(output_dict, indent=2)
        return json_output

model = GoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GOOGLE_API_KEY)
prompt = PromptTemplate.from_template(
    "List exactly 10 countries that import {product}. "
    "Respond ONLY with numbered list like:\n"
    "1. Country\n"
    "2. Country\n"
    "No extra text, no intro, no explanation."
)
chain = prompt | model | LLMOutputToJsonParser()

output = chain.invoke({"product": "rice"})
print(output)
```

## Sequential Chains

```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough

llm = GoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GOOGLE_API_KEY)

first_prompt = ChatPromptTemplate.from_template(
    "What is the best name to describe a company that makes {product}? Return only the name, no explanation."
)

second_prompt = ChatPromptTemplate.from_template(
    "Write a 20 words description for the following company: {company_name}"
)

chain_one = first_prompt | llm
chain_two = second_prompt | llm

overall_chain = (
    {"company_name": chain_one}
    | RunnablePassthrough.assign(description=chain_two)
)

result = overall_chain.invoke({"product": "gaming laptop"})
print("Company name: ", result["company_name"])
print("Description: ", result["description"])
```

## Router chains

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough, RunnableBranch
from langchain_google_genai import GoogleGenerativeAI

llm = GoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GOOGLE_API_KEY, temperature=0.7)

physics_prompt = PromptTemplate.from_template(
    "You are a physics expert. Answer this physics question: {input}"
)

math_prompt = PromptTemplate.from_template(
    "You are a math expert. Answer this math question: {input}"
)

default_prompt = PromptTemplate.from_template(
    "Answer this question: {input}"
)

router_prompt = PromptTemplate.from_template(
    "Classify the topic of this question as one word - either 'physics', 'math' or 'other'.\n"
)

router_chain = router_prompt | llm

branch = RunnableBranch(
    (lambda x: "physics" in x["topic"].lower(), physics_prompt | llm),
    (lambda x: "math" in x["topic"].lower(), math_prompt | llm),
    default_prompt | llm
)

full_chain = (
    RunnablePassthrough.assign(topic=router_chain)
    | branch
)

result = full_chain.invoke({"input": "What is the speed of light?"})
print(result)
```

## Vector stores

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores import FAISS

raw_documents = WebBaseLoader('https://en.wikipedia.org/wiki/Apollo_13').load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
documents = text_splitter.split_documents(raw_documents)

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={"device": "cpu"}  # Force CPU to avoid CUDA sm_61 incompatibility
)

db = FAISS.from_documents(documents, embeddings)
```

```python
query = "When did appolo 13 mission happen"
docs = db.similarity_search(query)
print(docs[0].page_content)
```

```python
retriever = db.as_retriever()
retrieved_docs = retriever.invoke(
    "When did appolo 13 happen?"
)
print(retrieved_docs[0].page_content)
```

```python
from langchain_google_genai import GoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", google_api_key=GOOGLE_API_KEY)
from langchain_core.tools import tool

@tool
def reverse_string(input_string):
    """
    Reverses the input string.

    Args:
    input_string (str): The string to be reversed.

    Returns:
    str: The reversed string.
    """
    return input_string[::-1]

tools = [reverse_string]
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_classic.agents.format_scratchpad.tools import format_to_tool_messages
from langchain_classic.agents.output_parsers.tools import ToolsAgentOutputParser

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", google_api_key=GOOGLE_API_KEY)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a very powerful assistant, but bad at reversing a given word."),
    ("user", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

llm_with_tools = llm.bind_tools(tools)

agent = {
    "input": lambda x: x["input"],
    "agent_scratchpad": lambda x: format_to_tool_messages(x["intermediate_steps"])
} | prompt | llm_with_tools | ToolsAgentOutputParser()
```

```python
agent.invoke({
    "input": "what is the reverse of everest",
    "intermediate_steps": []
})
```

```python
from langchain_classic.agents import AgentExecutor
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
```

```python
agent_executor.invoke({"input": "what is the reverse of everest?"})
```
